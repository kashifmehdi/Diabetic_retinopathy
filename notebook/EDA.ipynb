```python
import pandas as pd  #type: ignore
import matplotlib.pyplot as plt   #type: ignore
import seaborn as sns   #type: ignore
from PIL import Image  #type: ignore
import os
```

```python
# Load the labels CSV
labels_df =pd.read_csv('C:\\Users\\scl\\Desktop\\Kashif\\AI_Projects\\Diabetic Retinopathy\\data\\train.csv')
```

```python
# Display the first five rows of the dataset
labels_df.head()
```

```python
# Plot label distribution
sns.countplot(x='diagnosis', data=labels_df)
plt.title("Distribution of Diagnoses")
plt.show()
```

```python
# Display a few sample images
sample_images = labels_df.sample(5)['id_code'].values
for img_name in sample_images:
  img_path = os.path.join('C:\\Users\\scl\\Desktop\\Kashif\\AI_Projects\\Diabetic Retinopathy\\data\\train_images', f'{img_name}.png')
  img = Image.open(img_path)
  plt.imshow(img)
  plt.axis('off')
  plt.show()
```
